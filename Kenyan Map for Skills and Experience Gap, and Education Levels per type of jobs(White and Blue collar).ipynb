```python
#Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
#Load data from your source
data = pd.read_excel("C:/Users/KARISA/Downloads/all variables.xlsx", index_col = "County")
data
```

```python
#Let's load the shape file
shapefile = gpd.read_file("C:/Users/KARISA/Downloads/Shapefile")
shapefile
```

```python
#Merge the data file with the shapefile that you loaded
merged = shapefile.merge(data, how = "left", left_on = "county", right_index = True)
merged
```

```python
#Drop all missing variables
merged = merged.dropna()
merged
```

```python
#Let's plot the map now
fig, ax =plt.subplots(1,1, figsize = (10,10))
merged.plot(column = "skill_gap", cmap = "Oranges", linewidth = 0.5, ax=ax, legend = True)

# Add title and adjust plot settings
plt.title('Skills Gap by Major Cities in Kenya')
ax.axis('off')

# Create a legend for county names
merged.apply(lambda x: ax.annotate(text=x['county'], xy=x['geometry'].centroid.coords[0], horizontalalignment='center', fontsize=8), axis=1)

# Show the plot
plt.show()
```
